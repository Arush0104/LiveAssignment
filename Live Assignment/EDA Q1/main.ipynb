```python
male = (df['GENDER'] != 'M').sum()
female = (df['GENDER']=='M').sum()
print(f"Males:{male} and females:{female}")
```

```python
plt.figure(figsize=(20,15))
for i,column in enumerate(df.columns):
    plt.subplot(5,5,i+1)
    plt.hist(df[column])
    plt.title(f'{column} Distribution')
plt.tight_layout()
```

```python
# Calculate skewness
skewness = df['AGE'].skew()

# Create a distribution plot with a rug plot to show skewness
sns.displot(df['AGE'], kde=True, rug=True)

# Add a title and label for skewness
plt.title(f"Distribution of AGE")
plt.xlabel('col1 values')
plt.ylabel('Density')
plt.grid(True)

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)
```

```python
df['AGE'].plot(kind='box')

# Add a title and labels (optional)
plt.title('Box Plot of col1')
plt.xlabel('col1')
plt.ylabel('Value')
plt.grid(True)

plt.show()
```
